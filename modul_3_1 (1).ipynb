```python
import re
import json
def get_urls():
  url_list = []
  url= "https://index.minfin.com.ua/ua/russian-invading/casualties/"
  html_doc= requests.get(url)
  html_doc.text
  soup = BeautifulSoup(html_doc.text, "html.parser")
  content= soup.select("div[class=ajaxmonth] h4[class=normal] a")

  prefix = "casualties/"
  # print (content)
  # prefix = "month.php?month="
  for link in content:
    url_1 = prefix+re.search(r"\d{4}-\d{2}", link["href"]).group()
    url_list.append ("/"+url_1)
    return url_list
    print (url_list)
    # print (url_1)
  # russian-invading/casualties/2024-02/
```

```python
url = 'https://quotes.toscrape.com/'
response = requests.get(url)
soup = BeautifulSoup (response.text)
quotes = soup.find_all('span', class_='text')
authors = soup.find_all('small', class_='author')
tags = soup.find_all('div', class_='tags')
for i in range(0, len(quotes)):
    print(quotes[i].text)
    print('--' + authors[i].text)
    tagsforquote = tags[i].find_all('a', class_='tag')
    for tagforquote in tagsforquote:
        print(tagforquote.text)
    # break
```
